```python
from pyspark.sql import SparkSession

df = (
    spark.read.format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load("/Volumes/finops_catalog_2/finops/raw/cloud_budget_2023_dataset.csv")
)


df = df.dropDuplicates()
```

```python
spark.sql("DROP TABLE IF EXISTS finops_catalog_2.finops.bronze_cloud_budget")

df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("finops_catalog_2.finops.bronze_cloud_budget")
```
