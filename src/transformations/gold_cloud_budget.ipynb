```python
from pyspark.sql.functions import *

silver_df = spark.read.table("finops_catalog_2.finops.silver_cloud_budget")
```

```python
gold_daily_cost = (
    silver_df
    .groupBy("date")
    .agg(
        round(sum("net_cost"), 2).alias("total_cost")
    )
)
```

```python

gold_daily_cost.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("finops_catalog_2.finops.gold_daily_cost")
```

```python
gold_business_unit_cost = (
    silver_df
    .groupBy("business_unit")
    .agg(
        round(sum("net_cost"), 2).alias("total_cost")
    )
)
```

```python

gold_business_unit_cost.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("finops_catalog_2.finops.gold_business_unit_cost")
```

```python
gold_service_cost = (
    silver_df
    .groupBy("service")
    .agg(
        round(sum("net_cost"), 2).alias("total_cost")
    )
    .orderBy(desc("total_cost"))
)
```

```python

gold_service_cost.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("finops_catalog_2.finops.gold_service_cost")
```

```python
gold_department_cost = (
    silver_df
    .groupBy("department")
    .agg(
        round(sum("net_cost"), 2).alias("total_cost")
    )
)
```

```python

gold_department_cost.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("finops_catalog_2.finops.gold_department_cost")
```

```python
gold_anomaly_features = silver_df.select(
    "date",
    "business_unit",
    "department",
    "service",
    "net_cost",
    "cloud_provider",
    "budget_status",
    "usage_quantity",
    "budget_amount",
    "budget_utilization_pct",
    "cost_variance_7d_pct",
    "cost_variance_30d_pct",
    "anomaly_score",
    "is_anomaly"
)
```

```python
spark.sql("DROP TABLE IF EXISTS finops_catalog_2.finops.gold_anomaly_features")

gold_anomaly_features.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("finops_catalog_2.finops.gold_anomaly_features")
```
